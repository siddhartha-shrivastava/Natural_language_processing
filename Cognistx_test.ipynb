## Now it's time to explore the text features. First let's look at the most popular words for sarcastic comments.We will throw away most common words, like 'a', 'the', 'and' etc. and will concider unigramms , bigramms and Trigram separately

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
vectorizer_1 = CountVectorizer(stop_words='english', ngram_range=(1, 1))
vectorizer_2 = CountVectorizer(stop_words='english', ngram_range=(2, 2))
vectorizer_3 = CountVectorizer(stop_words='english', ngram_range=(3, 3))
```

```python
def freq_words(vectorizer, data):
    X = vectorizer.fit_transform(data)
    freqs = zip(vectorizer.get_feature_names(), np.asarray(X.sum(axis=0)).ravel())
    return sorted(freqs, key = lambda x: x[1], reverse=True)[:10]
```

```python
# OUTPUT ----> [(sarcastic comments,freq ), (non-sarcastic comments, freq)]
list_com = [freq_words(vectorizer_1, EDA_train[EDA_train['label']==1]['comment']),
     freq_words(vectorizer_1, EDA_train[EDA_train['label']==0]['comment'])]
list(map(list, zip(*list_com)))
```

```python
#lets try bigram
```

```python
# OUTPUT ----> [(sarcastic comments,freq ), (non-sarcastic comments, freq)]

list_com = [freq_words(vectorizer_2, EDA_train[EDA_train['label']==1]['comment']),
     freq_words(vectorizer_2, EDA_train[EDA_train['label']==0]['comment'])]
list(map(list, zip(*list_com)))
```

```python
#lets try trigram
# OUTPUT ----> [(sarcastic comments,freq ), (non-sarcastic comments, freq)]

list_com = [freq_words(vectorizer_3, EDA_train[EDA_train['label']==1]['comment']),
     freq_words(vectorizer_3, EDA_train[EDA_train['label']==0]['comment'])]
list(map(list, zip(*list_com)))
```
